# Sélectionner des données à partir des types

```
# packages
import pandas as pd
```

## read data

```
# data
orders_data = pd.read_csv("/datasets/google-drive/Datasets/orders.csv")
orders_data["order_date"] = pd.to_datetime(orders_data["order_date"], format="%Y/%m/%d")

print(orders_data.head())
```

```
# get columns data types
orders_data.info()
```

## Select some columns types

```
# select columns with int and float data types
print(orders_data.select_dtypes(include=["int", "float"]))
```

## Exclude some columns types

```
# exclude columns with int data type
print(orders_data.select_dtypes(exclude=["int"]))
```
