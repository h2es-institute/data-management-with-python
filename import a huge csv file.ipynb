```python
import pandas as pd
import numpy as np
```

```python
# fichier stocks des entreprises
filename_stock_etablissement = "datasets/StockEtablissement_utf8.csv"
```

```python
reader = pd.read_csv(filename_stock_etablissement, header=0, iterator=True, low_memory=False)
chunks = []
loop = True
while loop:
    try:
        chunk = reader.get_chunk(100000)
        chunks.append(chunk)
    except StopIteration:
        loop = False
        print("Iteration is stopped")

df_ac = pd.concat(chunks, ignore_index=True)
```

```python
TextFileReader = pd.read_csv(filename_stock_etablissement, chunksize=100000, low_memory=False)  # the number of rows per chunk

dfList = []
for df in TextFileReader:
    dfList.append(df)

df = pd.concat(dfList,sort=False)
```
