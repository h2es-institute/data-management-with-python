## `converters`

```python
# lecture de la troisième feuille
df = pd.read_excel(io = "datasets/les-arbres-plantes.xlsx",
                   sheet_name = "Sheet3", 
                   header = 3, 
                   converters = {"Emplacement - Identifiant unique": str}
                  )
# entête
df.head(2)
```

```python
df.info()
```

## `nrows`

```python
# lecture de la troisième feuille
df = pd.read_excel(io = "datasets/les-arbres-plantes.xlsx",
                   sheet_name = "Sheet3", 
                   header = 0, 
                   converters = {"Emplacement - Identifiant unique": str}, 
                   nrows=3
                  )
# entête
df.shape
```

## `skiprows`

```python
# lecture de la troisième feuille
df = pd.read_excel(io = "datasets/les-arbres-plantes.xlsx",
                   sheet_name = "Sheet1", 
                   header = 3, 
                   converters = {"Emplacement - Identifiant unique": str}, 
                   skiprows=lambda x: x%2==0
                  )
# shape
df.shape
```

## `true_values, false_values, na_values`

```python
# lecture de la troisième feuille
df = pd.read_excel(io = "datasets/les-arbres-plantes.xlsx",
                   sheet_name = "Sheet3", 
                   header = 3, 
                   converters = {"Emplacement - Identifiant unique": str}, 
                   true_values = "V",
                   false_values = "F",
                   na_values = "non renseigné"
                   
                  )
# entête
df
```

## `dtype`

```python
# lecture de la troisième feuille
df = pd.read_excel(io = "datasets/les-arbres-plantes.xlsx",
                   sheet_name = "Sheet3", 
                   header = 3, 
                   # converters = {"Emplacement - Identifiant unique": str}, 
                   true_values = "V",
                   false_values = "F",
                   na_values = "non renseigné",
                   dtype=str
                  )
# entête
df.info()
```

## `index_col`

```python
# lecture de la troisième feuille
df = pd.read_excel(io = "datasets/les-arbres-plantes.xlsx",
                   sheet_name = "Sheet3", 
                   header = 3, 
                   # converters = {"Emplacement - Identifiant unique": str}, 
                   true_values = "V",
                   false_values = "F",
                   na_values = "non renseigné",
                   dtype=str,
                   index_col = "Emplacement - Identifiant unique"
                  )
# entête
df.head()
```
