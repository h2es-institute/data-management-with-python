## header

```python
df = pd.read_csv(filepath_or_buffer="datasets/StockEtablissement_utf8.csv",
                 nrows=1000,
                 # sep = ",",
                 delimiter=",",
                 header=0
                 )
df.head(2)
```

## index_col

```python
df = pd.read_csv(filepath_or_buffer="datasets/StockEtablissement_utf8.csv",
                 nrows=1000,
                 sep = ",",
                 # delimiter=",",
                 header=0,
                 index_col = "siret"
                 )
df.head(2)
```

## converters

```python
df = pd.read_csv(filepath_or_buffer="datasets/StockEtablissement_utf8.csv",
                 nrows=1000,
                 sep = ",",
                 # delimiter=",",
                 header=0,
                 index_col = "siret", 
                 converters = {"siren": str, "siret": str}
                 )
df.head(2)
```

## skiprows

```python
df = pd.read_csv(filepath_or_buffer="datasets/StockEtablissement_utf8.csv",
                 nrows=1000,
                 sep = ",",
                 # delimiter=",",
                 header=0,
                 # index_col = "siret", 
                 converters = {"siren": str, "siret": str},
                 skiprows = 2
                 )
df.head(2)
```

# skip_blank_lines

```python
df = pd.read_csv(filepath_or_buffer="datasets/StockEtablissement_utf8.csv",
                 nrows=1000,
                 sep = ",",
                 # delimiter=",",
                 header=0,
                 # index_col = "siret", 
                 converters = {"siren": str, "siret": str},
                 skiprows = None,
                 skip_blank_lines = False
                 )
df.head(2)
```

## decimal

```python
df = pd.read_csv(filepath_or_buffer="datasets/StockEtablissement_utf8.csv",
                 nrows=1000,
                 sep = ",",
                 # delimiter=",",
                 header=0,
                 # index_col = "siret", 
                 converters = {"siren": str, "siret": str},
                 skiprows = None,
                 skip_blank_lines = False,
                 decimal = "."  # indicateur de la partie décimale
                 )
df.head(2)
```

## encoding

```python
df = pd.read_csv(filepath_or_buffer="datasets/StockEtablissement_utf8.csv",
                 nrows=1000,
                 sep = ",",
                 # delimiter=",",
                 header=0,
                 # index_col = "siret", 
                 converters = {"siren": str, "siret": str},
                 skiprows = None,
                 skip_blank_lines = False, 
                 encoding = "utf-8"
                 )
df.head(2)
```

## on_bad_lines

```python
df = pd.read_csv(filepath_or_buffer="datasets/StockEtablissement_utf8.csv",
                 nrows=1000,
                 sep = ",",
                 # delimiter=",",
                 header=0,
                 # index_col = "siret", 
                 converters = {"siren": str, "siret": str},
                 skiprows = None,
                 skip_blank_lines = False,
                 on_bad_lines = "skip"  # ('warn', 'error')ne pas lire les lignes qui ont un pb (notamment s'il y a trop ou moins de colonnes)
                 )
df.head(2)
```

## low_memory

```python
df = pd.read_csv(filepath_or_buffer="datasets/StockEtablissement_utf8.csv",
                 nrows=10000,
                 sep = ",",
                 # delimiter=",",
                 header=0,
                 # index_col = "siret", 
                 converters = {"siren": str, "siret": str},
                 skiprows = None,
                 skip_blank_lines = False,
                 on_bad_lines = "skip",  # ('warn', 'error')ne pas lire les lignes qui ont un pb (notamment s'il y a trop ou moins de colonnes)
                 low_memory=False 
                )
df.head(2)

# Pour éviter des types mixtes, il faut spécifier low_memory à False ou bien spécifier le type de la colonne
```
