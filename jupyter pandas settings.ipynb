# Modifier les paramètres par défaut d'affichage de dataframe sur Jupyter Notebook

```python
import pandas as pd
import numpy as np
```

# Nombre maximal de lignes

```python
# Nombre de lignes par défaut
print(pd.get_option('display.max_rows'))
print(pd.options.display.max_rows)
```

```python
df = pd.DataFrame(np.random.randn(100, 3))
df
```

```python
# modifier la valeur
pd.set_option('display.max_rows', 400)
```

```python
df
```

# Nombre maximal de colonnes

```python
# Nombre de colonnes par défaut
print(pd.get_option('display.max_columns'))
print(pd.options.display.max_columns)
```

```python
df = pd.DataFrame(np.random.randn(10, 30))
df
```

```python
# modification du nombre maximal de colonnes à afficher
pd.options.display.max_columns = 50
```

```python
df
```

# Largeur de colonnes

```python
# Largeur de colonnes
print(pd.get_option('max_colwidth'))
print(pd.options.display.max_colwidth)
```

```python
df = pd.DataFrame({"text": ["pandas jupyter settings "*2, "print data into notebook jupyter"*3],
                   "range": [1, 2]
                  })
df
```

```python
pd.options.display.max_colwidth = 100
```

```python
df
```

# Nombre de colonnes à afficher avec la méthode `info()`

```python
# Nombre maximal de colonnes par défaut 
pd.options.display.max_info_columns
```

```python
# changeons la valeur
pd.options.display.max_info_columns = 10
```

```python
df = pd.DataFrame(np.random.randn(10, 30))
df.info()
```

```python
# changeons la valeur
pd.options.display.max_info_columns = 10
df = pd.DataFrame(np.random.randn(10, 9))
df.info()
```

# Nombre de décimales

```python
# nombre de chiffres à afficher dans la partie décimale
pd.options.display.precision
```

```python
df = pd.DataFrame(10+np.random.randn(10, 3))
df
```

```python
# modifions à deux chiffres dans la partie décimale
pd.options.display.precision=2

# affichons à nouveau le df
df
```
