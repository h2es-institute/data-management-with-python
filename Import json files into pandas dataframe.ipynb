# Lire un fichier json et le convertir en dataframe python

```python
# packages
import pandas as pd
import json 

# options
pd.set_option("display.max_columns", 200)

# chemin ver le fichier json
json_file_path = "datasets/jsonfilewithhierarchy-100.json"
```

## Lecture avec `json` et conversion en dataframe

```python
# lecture du fichier
with open(json_file_path, encoding="latin-1") as json_file:
    json_data = json.load(json_file)
    json_file.close()
```

```python
# cl√©s du fichier
json_data.keys()
```

```python
# √©chantillon fichier
json_data
```

```python
# Cr√©ation du dataframe avec l'objet json
df1 = pd.DataFrame(json_data['feeds'])

# ent√™te du dataframe
df1.head()
```

```python
# convertion de la colonne 'feeds' en plusieurs colonnes dans le dataframe
df2 = pd.json_normalize(json_data["feeds"], max_level=10).head(10)
df2.head()
```

## Utiliation du la fonction `read_json()` de `pandas`

```python
# utiliation du la foncion
df_pd = pd.read_json(path_or_buf=json_file_path)
df_pd.head()
```

```python
# convertion du dictionnaire dans 'feeds' en colonne du dataframe
df_pd["feeds"].apply(pd.Series).head()
```
